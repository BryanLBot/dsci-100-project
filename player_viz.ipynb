```r
library(tidyverse)
library(cowplot)
library(scales)
library(repr)
library(ggplot2)

players <- read_csv("data/players.csv")
players
```

```r
player_AgeVsHours_plot <- players |>
                        ggplot(aes(x = age, y = played_hours)) +
                        geom_point() +
                        xlab("Player Age (years)") +
                        ylab("Played hours")


player_AgeVsHours_plot
```
