```r
# Load the datsets
sessions <- read.csv("data/sessions.csv")
players <- read.csv("data/players.csv")
```

```r
# Check out the datasets
head(sessions)
head(players)
```

```r
# Since sessions is not a tidy dataset, we need to fix it up
sessions <- sessions |>
    mutate(
        start_time = parse_date_time(start_time, orders = "dmy HM"),
        end_time = parse_date_time(end_time, orders = "dmy HM")
    )

sessions <- sessions |>
    mutate(duration = as.numeric(difftime(end_time, start_time, units = "mins")))

sessions <- sessions |>
    separate(col = start_time, into = c("start_date", "start_time"), sep = " ", fill = "right")

sessions <- sessions |>
    separate(col = end_time, into = c("end_date", "end_time"), sep = " ", fill = "right")

sessions <- sessions |>
    filter(!is.na(start_date), !is.na(start_time), !is.na(end_date), !is.na(end_time))

head(sessions)
```

```r
# Same for players, since the dataset is not tidy, we have to make it into tidy data.

players <- players |>
    select(-individualId, -organizationName)

head(players)
```

```r
sessions <- select(sessions, hashedEmail, start_date, start_time, end_date, end_time, duration)
head(sessions)
head(players)
```

**Merging and Summarizing data:**

To analyze the relationships between player traits and their total gameplay metrics, I merge players and sessions datasets. I will calculate key metrics such as total play duration and session count per player.

```r
# Summarize sessions data
sessions_summary <- sessions |>
    group_by(hashedEmail) |>
    summarize(
        total_sessions = n(),
        total_duration = sum(duration, na.rm = TRUE),
        avg_duration = mean(duration, na.rm = TRUE),
    )

# Merge with players data
player_analysis <- players |>
    inner_join(sessions_summary, by = "hashedEmail")

# Look at the merged data
head(player_analysis)
# player_analysis: 125 rows
```

# Methods and Plans:

To address the question of "What traits of players lead to more playtime?" the primary method will be k-Nearest Neighbors (k-NN), with Linear Regression used as a supplementary method to provide interpretability and validate the results.

**Why are these methods appropriate?**

K-NN is well-suited for identifying patterns in non-linear data and making predictions about total playtime based on similar players. And by adjusting the number of neighbors (k) and the distance metric, K-NN can flexibly model complex relationships between player traits and playtime.
Linear regression provides interpretable coefficients that quantify how each trait influences playtime in a linear approximation. It serves as a benchmark to validate whether K-NN captures meaningful patterns.

**Assumptions**

The k-Nearest Neighbors method assume that similar players (neighbors) in the feature space will have similar playtimes. It will require the features to be scaled and encoded appropriately for meaningful distance calulations. Whlie the lienar regression method assumes a linear relationship between predictors and playtime.

**Limitations or Weaknesses**

k-NN's performace depends heavily on the choice of k. A low k leads to overfitting, while a high k results in underfitting. It is also computationally expensive for large datasets due to the need to calculate distances for each prediction.

Linear regression is limited to lienar relationships and sensitive to outliters. So it doesn't handle non-linear or complex patterns without significant preprocessing.

**How to compare and select the model**

**For k-NN:**
  - Use cross-validation to tune k, testing a range of values to find the one that minimizes error metrics like the **root mean squared error (RMSE)**.
  - Evaluate performance on the testing set to ensure generalization.

**For Linear Regression:**
  - Similar to k-NN, evaluate the model using the error in prediction (looking at RMSE and trying to minimize it).
  - Use regression coefficients to identify the magnitude and direction of influence for each trait.

**Data Processing and Validation Plan**

**1. Data Splitting:**
   - Split the dataset into training (70%) and testing (30%) sets.
   - Cross-validation will be used during training to ensure robust evaluation of k for k-NN and to avoid overfitting.

**2. Feature Engineering:**
   - Encode categorical variables (e.g., gender, experience) as numeric values.
   - Scale numerical features (e.g., age, total sessions) to ensure fair comparison in k-NN's distance calculations.

**3. Handling Outliers:**
   - Apply techniques like log transformations or robust scaling to reduce the influence of outliers in total playtime.

**Next Steps:**

1. Implement k-NN as the primary method to predict total playtime based on player traits, tuning k for optimal performance.

2. Use linear regression to provide a simple, interpretable model of the relationships, validating whether traits identified as significant by K-NN align with regression results.
